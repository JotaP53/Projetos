# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

### Minhas anotações

```python
# NOTAS SOBRE O PYAUTOGUI
# pyautogui.click -> Clicar com o mouse.
# pyautogui.write -> Escrever um texto.
# pyautogui.press -> Apertar alguma tecla.
# pyautogui.hotkey -> Apertar teclas combinadas (p.ex.: Ctrl + C)

# BIBLIOTECAS QUE SERÃO UTILIZADAS
# Pyautogui -> Para automações.
# Pandas - Para trabalhar com os dados.
# OpenPyXL -> Para permitir o Pandas trabalhar com arquivos do Excel.

# COMANDO PARA INSTALAR AS BIBLIOTECAS
# pip install pandas pyautogui openpyxl
```

### Importando Bibliotecas

```python
import pyautogui
import pyperclip
import time
import pandas
import openpyxl
```

### Passo 1: Entrar no Sistema da Empresa (Link do Drive)

```python
# Esse comando servirá para gerar uma pausa entre cada comando do pyautogui
pyautogui.PAUSE = 1
```

```python
# Abrir o Chrome.
pyautogui.press('Win')
pyautogui.write('Google Chrome')
pyautogui.press('Enter')

# Pausará a execução do código pela quantidade de segundos especificada.
time.sleep(2)

# Digitar o link do sistema.
link_drive = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga'
pyperclip.copy(link_drive)

# Apertar "Enter" e esperar.
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('Enter')
time.sleep(3)
```

### Passo 2: Navegar no sistema para encontrar a base de dados.

```python
# Comando de navegação.
pyautogui.click(x=549, y=483, clicks = 2)
time.sleep(1)
```

### Passo 3: Exportar a base de dados (baixar o arquivo).

```python
pyautogui.click(x=806, y=488)
time.sleep(1)
pyautogui.click(x=2347, y=482)
time.sleep(5)
pyautogui.press('Enter')
time.sleep(3)
```

### Passo 4: Calcular os indicadores (faturamento e quantidade de produtos vendidos).

- Faturamento
- Quantidade de Produtos

```python
# Abrir a base de dados
caminho = r'C:\Users\João Pedro\Downloads\Vendas - Dez.xlsx'
tabela = pandas.read_excel(caminho)

# Ver as informa;óes da base de dados.
# O "Display" é uma forma melhor que o "Print" de ler uma tabela, pois através dele a tabela vem mais formatada, com um layout mais amigável.
display(tabela)

# Somar o faturamento de todos os produtos (somar a coluna de Valor Final).
faturamento = tabela["Valor Final"].sum()
print(faturamento)

# Somar a quantidade de produtos (somar a coluna de Quantidade).
qtde_produtos = tabela["Quantidade"].sum()
print(qtde_produtos)
```

### Passo 5: Enviar as informações por e-mail.

```python
# Abrir uma nova aba.
pyautogui.hotkey('ctrl', 't')

# Pausará a execução do código pela quantidade de segundos especificada.
time.sleep(2)


# Digitar o link do meu e-mail.
link_email = 'https://mail.google.com/'
pyperclip.copy(link_email)

# Entrar no email.
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('Enter')
time.sleep(8)

# Clicar no botáo escrever.
pyautogui.click(x=72, y=216)
time.sleep(2)

# Para quem enviar o e-mail.
endereco_email = 'martinsjp53+projeto@gmail.com'
pyperclip.copy(endereco_email)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('Tab')
time.sleep(1)

# Qual o assunto do e-mail.
assunto_email = 'Relatório de Vendas'
pyperclip.copy(assunto_email)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('Tab')
time.sleep(1)

# Qual o corpo do e-mail.
texto = f'''
Prezados,

Segue relatório de vendas de ontem.

Faturamento: R$ {faturamento:,.2f}
Quandiade de Produtos Vendidos: R$ {qtde_produtos:,.2f}

Qualquer dúvida estou à disposição.

Abs,
'''

pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')

# Enviar o e-mmail.
pyautogui.click(x=1952, y=1005)
```
